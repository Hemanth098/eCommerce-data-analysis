```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv('Customers.csv')
products = pd.read_csv('Products.csv')
transactions = pd.read_csv('Transactions.csv')
```

```python
print(customers.info())
print(products.info())
print(transactions.info())
```

```python
data = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')
data.dropna(inplace=True)
data = data[data['TotalValue'] > 0]
# Summary statistics
print(data.describe())
print(data.info())
```

```python
# EDA visualizations
# 1. Distribution of transactions over time
plt.figure(figsize=(10, 5))
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'], errors='coerce')
data.dropna(subset=['TransactionDate'], inplace=True)
data.set_index('TransactionDate').resample('M').size().plot(kind='line', title='Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.show()

```

```python
# 2. Sales by region
plt.figure(figsize=(10, 5))
sales_by_region = data.groupby('Region')['TotalValue'].sum()
sales_by_region.plot(kind='bar', title='Sales by Region')
plt.show()
```

```python
# 3. Popular products
plt.figure(figsize=(10, 5))
popular_products = data['ProductName'].value_counts().head(10)
popular_products.plot(kind='bar', title='Top 10 Popular Products')
plt.show()
```

```python
# 4. Revenue by category
plt.figure(figsize=(10, 5))
revenue_by_category = data.groupby('Category')['TotalValue'].sum()
revenue_by_category.plot(kind='bar', title='Revenue by Category')
plt.show()
```

```python
# Task 2: Lookalike Model

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Creating a customer-product matrix
customer_product_matrix = data.pivot_table(index='CustomerID', columns='ProductName', values='Quantity', fill_value=0)
similarity_matrix = cosine_similarity(customer_product_matrix)

# Building a lookalike function
def find_lookalikes(customer_id, top_n=3):
    customer_idx = customer_product_matrix.index.get_loc(customer_id)
    similarity_scores = list(enumerate(similarity_matrix[customer_idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similar_customers = [(customer_product_matrix.index[i], score) for i, score in similarity_scores[1:top_n+1]]
    return similar_customers
```

```python
# Generating lookalikes for first 20 customers
lookalike_results = {}
for customer in customers['CustomerID'][:20]:
    lookalike_results[customer] = find_lookalikes(customer)

# Saving results to CSV
lookalike_df = pd.DataFrame.from_dict(lookalike_results, orient='index', columns=['Lookalike 1', 'Lookalike 2', 'Lookalike 3'])
lookalike_df.to_csv('Lookalike.csv', index_label='CustomerID')
lookalike_df
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Feature engineering
features = data.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'TransactionID': 'count',
    'Price_x': 'mean',
    'Price_y': 'mean',
    'Quantity': 'mean'
}).reset_index()
features.fillna(0, inplace=True)
features = features[features['TotalValue'] > 0]
features
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features.iloc[:, 1:])
scaled_features
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_features)
features['Cluster'] = kmeans.labels_
kmeans
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=features['TotalValue'], y=features['TransactionID'], hue=features['Cluster'], palette='Set2')
plt.title('Customer Segmentation')
plt.xlabel('Total Value')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score

print("DB Index:", davies_bouldin_score(scaled_features, kmeans.labels_))

# Save cluster assignments
features.to_csv('Customer_Clusters.csv', index=False)
```
